```python
startups=startups.rename({'R&D Spend':'RDS','Administration':'ADM','Marketing Spend':'MKS'},axis=1)
startups
```

```python
import statsmodels.formula.api as smf 
model = smf.ols('Profit~RDS+ADM+MKS',data=startups).fit()
```

```python
model.params
```

```python
print(model.tvalues, '\n', model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
p_a=smf.ols('Profit~ADM',data = startups).fit()  
print(p_a.tvalues, '\n', p_a.pvalues)  
```

```python
p_a=smf.ols('Profit~MKS',data = startups).fit()  
print(p_a.tvalues, '\n', p_a.pvalues)  
```

```python
p_am=smf.ols('Profit~ADM+MKS',data = startups).fit()  
print(p_am.tvalues, '\n', p_am.pvalues)
```

```python
rsq_r=smf.ols("RDS~ADM+MKS",data=startups).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADM~RDS+MKS",data=startups).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MKS~RDS+ADM",data=startups).fit().rsquared
vif_m=1/(1-rsq_m)

d1={'Variables':['RDS','ADM','MKS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') 
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "RDS", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "ADM", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "MKS", fig=fig)
plt.show()
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(startups)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
