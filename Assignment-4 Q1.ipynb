```python
import sklearn
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf
from sklearn.metrics import r2_score
```

```python
dtime = pd.read_csv("delivery_time.csv")
```

```python
dtime
```

```python
dtime=dtime.rename({'Delivery Time':'delivery_time', 'Sorting Time':'sorting_time'},axis=1)
```

```python
first_model = smf.ols("delivery_time~sorting_time",data=dtime).fit()
```

```python
first_model.params
```

```python
first_model.summary()
```

```python
model2 = smf.ols('delivery_time~np.log(sorting_time)',data=dtime).fit()
model2.summary()
```

```python
model3 = smf.ols('np.log(delivery_time)~sorting_time',data=dtime).fit()
model3.summary() 
```

```python
dtime["sorting_time_Sq"] = dtime.sorting_time**2
model_quad = smf.ols("np.log(delivery_time)~sorting_time+sorting_time_Sq",data=dtime).fit()
```

```python
model_quad.summary()
```

```python
np.exp(model_quad.predict(dtime[["sorting_time","sorting_time_Sq"]]))
```
